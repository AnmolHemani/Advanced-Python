```python
# Data Processing
transcations=[]
for i in range(0,7501):
    transcations.append([str(df.values[i,j]) for j in range(0,20)])

# Training the dataset
from apyori import apriori
rules = apriori(transcations,
                 min_support=0.003, #(3items*7 days in a week)/(Total rows in dataset 7501)
                   min_confidence=0.2,
                     min_left=3,
                       min_length=2)

# Visualizing the results
a = list(rules)
result=[list(a[i][0]) for i in range(0,len(a))]
```
