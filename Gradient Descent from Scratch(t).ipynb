```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4, n_features=1,
                      n_informative=1,
                       n_targets=1,
                         noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True,
                n_jobs=None)
```

```python
reg.coef_   # m slope
```

```python
reg.intercept_ # b intercept
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red')
```

```python
# Let's apply Gradient Descent assuming slope is 
# constant = 78.35
# and let's assume the starting value for 
# intercept is b=0

y_pred = ((78.35 * x) + 0).reshape(4) #y=mx+b
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')
plt.plot(x, y_pred, color='green', label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_slope = -2*np.sum(y-m*x.ravel()-b) #dL/db
loss_slope
```

```python
# Let's take learning rate=0.1

lr=0.1
step_size = lr * loss_slope
step_size
```

```python
# Calculating the new intercept

b = b - step_size # b_new=b_old-learning_rate*dL/db
b
```

```python
y_pred1 = ((78.35*x)+b).reshape(4)
y_pred1
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='black', label='OLS')
plt.plot(x, y_pred1, color='red', label='b={}'.format(b))
plt.plot(x, y_pred, color='yellow', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2

loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35*x)+b).reshape(4)
y_pred2
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')
plt.plot(x, y_pred2, color='green', label='b={}'.format(b))
plt.plot(x, y_pred, color='yellow', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 3

loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = lr * loss_slope 
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35*x)+b).reshape(4)
y_pred3
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color='red', label='OLS')
plt.plot(x, y_pred3, color='brown', label='OLS')
plt.plot(x, y_pred2, color='green', label='b={}'.format(b))
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))
plt.plot(x,y_pred, color='yellow', label='b=0')
plt.legend()
plt.show()
```
