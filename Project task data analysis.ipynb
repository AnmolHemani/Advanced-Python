```python
plt.figure(figsize=(10, 6))
numerical_cols = df.select_dtypes(include='number')
sns.heatmap(numerical_cols.corr())
plt.title('Correlation Between Numerical Variables')
plt.show()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('standard', StandardScaler(), numerical_cols),
        ('normalize', MinMaxScaler(), numerical_cols),
        ('encode', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ]
)
```

```python
transformed_data = preprocessor.fit_transform(df)
```

```python
standardized_data = preprocessor.named_transformers_['standard'].transform(df[numerical_cols])
normalized_data = preprocessor.named_transformers_['normalize'].transform(df[numerical_cols])
encoded_data = preprocessor.named_transformers_['encode'].transform(df[categorical_cols])
```

```python
standardized_df = pd.DataFrame(
    standardized_data,
    columns=numerical_cols.to_series().add_suffix('_std').values
)

normalized_df = pd.DataFrame(
    normalized_data,
    columns=numerical_cols.to_series().add_suffix('_norm').values
)

encoded_df = pd.DataFrame(
    encoded_data,
    columns=preprocessor.named_transformers_['encode'].get_feature_names_out(categorical_cols)
)
```

```python
final_transformed_df = pd.concat([standardized_df, normalized_df, encoded_df], axis=1)
```
