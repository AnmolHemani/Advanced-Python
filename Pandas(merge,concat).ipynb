```python
import numpy as np
import pandas as pd
```

```python
# Merge and concat two dataframes in pandas
```

```python
data1 = {
  "name": ["anmol", "gursewak", "prakul"],
  "age": [23,22,20],
  "Stream" : ["Data Science","Data Analyst","Python Developer"]
}

data2 = {
  "name": ["Ansul", "Heeman", "Vineet"],
  "age": [23,22,24],
  "Stream" : ["AWS","Djanngo","Python developer"]
}
```

```python
df1=pd.DataFrame(data1)
df1
```

```python
df2=pd.DataFrame(data2)
df2
```

```python
new_df=df1.merge(df2,how='outer')
new_df
```

```python
new_df=df1.merge(df2,how='left')
new_df
```

```python
new_df=df1.merge(df2,how='right')
new_df
```

```python
new_df=df1.merge(df2,how='cross')
new_df
```

```python
a=[df1,df2]
new_df1=pd.concat(a)
new_df1.reset_index()
```

```python
result1 = pd.concat([df1, df2], axis=1)
result1
```

```python
data3 = {
    "id": [101, 102, 103],  
    "name": ["anmol", "gursewak", "prakul"],
    "age": [23, 22, 20],
    "Stream": ["Data Science", "Data Analyst", "Python Developer"]
}

data4 = {
    "id": [101,102,103],  
    "project":["Supply Chain","Attrition","App sentiment analysis"]
}
```
