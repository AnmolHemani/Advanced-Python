```python
country_counts = df['Country'].value_counts()
country_counts
```

```python
# Distribution of entries by country.
    # Analyse which country adopted ai.
    # Count the number of entries per country and visualize with a bar chart.

plt.figure(figsize=(12,6))
sns.barplot(x=country_counts.index, y=country_counts.values, color='green')
plt.title('Distribution of Entries by Country')
plt.xlabel('Country')
plt.ylabel('Number of Entries')
plt.tight_layout()
plt.show()
```

```python
yearly_trends = df.groupby('Year')[['AI Adoption Rate (%)', 'Job Loss Due to AI (%)']].mean()
yearly_trends
```

```python
year_counts = df['Year'].value_counts()
year_counts
```

```python
# Number of entries per year

plt.figure(figsize=(12, 6))
sns.barplot(x=year_counts.index, y=year_counts.values, color='blue')
plt.title('Number of Entries per Year')
plt.xlabel('Year')
plt.ylabel('Entry Count')
plt.tight_layout()
plt.show()
```

```python
# AI Adoption Rate Over Time

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_trends, x='Year', y='AI Adoption Rate (%)', marker='o', label='AI Adoption Rate')
plt.title('AI Adoption Rate Over Time (2020–2025)')
plt.xlabel('Year')
plt.ylabel('Average AI Adoption Rate (%)')
plt.tight_layout()
plt.show()
```

```python
industry_metrics = df.groupby('Industry')[['AI Adoption Rate (%)', 'Job Loss Due to AI (%)', 'Revenue Increase Due to AI (%)']].mean()
industry_metrics
```

```python
# # AI Adoption Rate by Industry

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Industry', y='AI Adoption Rate (%)',hue='Industry', palette='Blues')
plt.title('AI Adoption Rate by Industry')
plt.tight_layout()
plt.show()
```

```python
# Job Loss Due to AI by Industry

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Industry', y='Job Loss Due to AI (%)',hue='Industry', palette='Oranges')
plt.title('Job Loss Due to AI by Industry')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Revenue Increase Due to AI by Industry

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Industry', y='Revenue Increase Due to AI (%)',hue='Industry', palette='Greens')
plt.title('Revenue Increase Due to AI by Industry')
plt.tight_layout()
plt.show()
```

```python
industry_content = df.groupby('Industry')['AI-Generated Content Volume (TBs per year)'].mean()
industry_content
```

```python
# Average Content Volume by Industry

plt.figure(figsize=(12, 6))
sns.barplot(x=industry_content.index, y=industry_content.values,hue=industry_content.index, palette='Oranges')
plt.title("Average AI-Generated Content Volume by Industry")
plt.ylabel("Average Content Volume (TBs/year)")
plt.tight_layout()
plt.show()
```

```python
distribution = df.groupby('Regulation Status')['Human-AI Collaboration Rate (%)'].mean()
distribution
```

```python
# Human-AI Collaboration Rate by Regulation Status

plt.figure(figsize=(10, 5))
sns.barplot(x=distribution.index, y=distribution.values,hue=distribution.index, palette='Blues')
plt.title("Human-AI Collaboration Rate by Regulation Status")
plt.xlabel("Regulation Status")
plt.ylabel("Collaboration Rate (%)")
plt.tight_layout()
plt.show()
```

```python
tool_counts = df['Top AI Tools Used'].value_counts()
tool_counts
```

```python
# Top AI Tools Frequency Across All Entries

plt.figure(figsize=(12, 6))
sns.barplot(x=tool_counts.index, y=tool_counts.values, color='blue')
plt.title('Top AI Tools Frequency Across All Entries')
plt.xlabel('AI Tool')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
c_trust = df.groupby('Consumer Trust in AI (%)')['AI Adoption Rate (%)'].mean()
c_trust
```

```python
# Consumer Trust vs. AI Adoption Rate

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Consumer Trust in AI (%)', y='AI Adoption Rate (%)', color = 'blue')
plt.title("Consumer Trust vs. AI Adoption Rate")
plt.xlabel("Consumer Trust in AI (%)")
plt.ylabel("AI Adoption Rate (%)")
plt.tight_layout()
plt.show()
```

```python
market_share_by_country = df.groupby('Country')['Market Share of AI Companies (%)'].mean()
market_share_by_country
```

```python
# Average Market Share of AI Companies by Country

plt.figure(figsize=(12, 6))
sns.barplot(x=market_share_by_country.index, y=market_share_by_country.values, color='Yellow')
plt.title("Average Market Share of AI Companies by Country")
plt.xlabel("Country")
plt.ylabel("Average Market Share (%)")
plt.tight_layout()
plt.show()
```
